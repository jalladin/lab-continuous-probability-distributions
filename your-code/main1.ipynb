    * Debe aceptar 3 parámetros: 
        * `bottom` - el límite inferior de los números generados
        * `ceiling` - el límite superior de los números generados
        * `count` - cuántos números generar
    * Debe devolver una matriz de números aleatorios distribuidos uniformemente

2. Llama a tu función con 2 conjuntos de parámetros a continuación:
    * bottom=10, ceiling=15, count=100
    * bottom=10, ceiling=60, count=1,000

3. Traza las distribuciones uniformes generadas anteriormente utilizando histogramas, donde el eje x es el valor y el eje y es el recuento. Que el número de intervalos del histograma sea 10.

El resultado debe ser como el siguiente:



```python
def generar_numeros_uniformes(bottom, ceiling, count):
    return uniform.rvs(loc=bottom, scale=ceiling-bottom, size=count)

numeros1 = generar_numeros_uniformes(10, 15, 100)
numeros2 = generar_numeros_uniformes(10, 60, 1000)


fig, axs = plt.subplots(2, 1, figsize=(10, 8))
axs[0].hist(numeros1, bins=10, edgecolor='black')# Primer histograma
axs[0].set_title('Distribución Uniforme: bottom=10, ceiling=15, count=100')
axs[0].set_xlabel('Valor')
axs[0].set_ylabel('Recuento')


axs[1].hist(numeros2, bins=10, edgecolor='black')# Segundo histograma
axs[1].set_title('Distribución Uniforme: bottom=10, ceiling=60, count=1000')
axs[1].set_xlabel('Valor')
axs[1].set_ylabel('Recuento')

plt.tight_layout()
plt.show()
```

How are the two distributions different?

Range:

First Distribution: Values range from 10 to 15.

Second Distribution: Values range from 10 to 60.

Number of Values:

First Distribution: 100 values.

Second Distribution: 1,000 values

## Distribución Exponencial

1. Utilizando `numpy.random.exponential`, crea una función que devuelva una lista de números distribuidos exponencialmente con la media de 10. 

1. Utiliza la función para generar dos secuencias de números de tamaño 10 y 100.

1. 2. Traza las distribuciones como histogramas con el número de bins como 100.

El resultado debe ser como el siguiente:



```python
def generar_exponencial(media, tamaño):
    lambda_ = 1 / media
    return np.random.exponential(scale=1/lambda_, size=tamaño)
```

```python
numeros_10 = generar_exponencial(10, 10)
numeros_100 = generar_exponencial(10, 100)

fig, axs = plt.subplots(2, 1, figsize=(10, 8))

axs[0].hist(numeros_10, bins=100, edgecolor='black', alpha=0.7)
axs[0].set_title('Distribución Exponencial (tamaño=10)')
axs[0].set_xlabel('Valor')
axs[0].set_ylabel('Frecuencia')

axs[1].hist(numeros_100, bins=100, edgecolor='black', alpha=0.7)
axs[1].set_title('Distribución Exponencial (tamaño=100)')
axs[1].set_xlabel('Valor')
axs[1].set_ylabel('Frecuencia')

plt.tight_layout()
plt.show()
```

¿En qué se diferencian ambas distribuciones?

Las dos distribuciones generadas con tamaños de muestra 10 y 100 difieren principalmente en suavidad y precisión. Con una muestra de tamaño 10, el histograma muestra una forma más irregular y menos representativa de la distribución exponencial, mientras que con una muestra de tamaño 100, el histograma se aproxima más a la forma teórica de la distribución exponencial, mostrando una curva más suave y continua. La muestra más grande proporciona una representación más precisa y menos variable de la distribución real.

La media cambia, por lo que la distribución también cambia. 

## Distribución Exponencial de datos reales

Supongamos que el tiempo que uno pasa en un banco se distribuye exponencialmente con media de 10 minutos (es decir, λ = 1/10). ¿Cuál es la probabilidad de que un cliente pase menos de quince minutos en el banco? 

Escribe un código en python para resolver este problema

```python
media = 10  # minutos
lambda_ = 1 / media
tiempo = 15  # minutos

probabilidad = expon.cdf(tiempo, scale=media)

print(f"La probabilidad de que un cliente pase menos de {tiempo} minutos en el banco es {probabilidad:.4f}")
```

¿Cuál es la probabilidad de que el cliente pase más de 15 minutos?

```python
probabilidad_mas = 1 - expon.cdf(tiempo, scale=media)

print(f"La probabilidad de que un cliente pase más de {tiempo} minutos en el banco es {probabilidad_mas:.4f}")
```
